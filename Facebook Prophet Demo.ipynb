```python
from fbprophet import Prophet
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/gregorychase/Downloads/prophet-master/examples/example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods = 365)
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
